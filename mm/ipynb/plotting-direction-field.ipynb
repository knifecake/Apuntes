```python
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')


samples = 15
z1, z2 = np.meshgrid(np.linspace(0, 2, samples), np.linspace(0, 2, samples))

# Lotka-Volterra system
a = 0.5
f1 = (1 - z2)*z1
f2 = a*(z1 - 1)*z2

plt.quiver(z1, z2, f1, f2)
plt.show()
```
